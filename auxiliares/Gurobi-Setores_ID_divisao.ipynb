```python
import pandas as pd

df = pd.read_csv('Documents/Agregados_por_setores_basico_BR_20250417.csv', sep=';')

# Preenche vazios com um ponto '.' (igual ao seu exemplo anterior) antes de juntar
bairro_str = df['CD_BAIRRO'].fillna('.').astype(str)
distrito_str = df['CD_DIST'].astype(str)

# Criando a nova coluna "ID_divisao"
df['ID_divisao'] = distrito_str + '_' + bairro_str

# Visualizando o resultado
print(df[['CD_DIST', 'CD_BAIRRO', 'ID_divisao']].head())
```

```python
# Salvar o DataFrame em um arquivo CSV
df.to_csv('SP_setores_populacao.csv', index=False, encoding='utf-8-sig', sep=';')
```
