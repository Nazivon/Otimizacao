```python
import pandas as pd

# 1. Carregar os arquivos
# Substitua pelos nomes reais dos seus arquivos
df_facilidades = pd.read_csv('Documents/Hospitais_gerais_SP_CNES.csv')
df_datasus = pd.read_csv('Downloads/TAB415/Dados/STSP2512.csv', encoding='latin1')

# --- ETAPA 1: Limpar o CNES (Remover zeros à esquerda) ---
# Convertendo para string e removendo os zeros à esquerda para garantir o match
df_facilidades['CO_CNES'] = df_facilidades['CO_CNES'].astype(str).str.lstrip('0')
df_datasus['CO_CNES'] = df_datasus['CNES'].astype(str).str.lstrip('0')

# --- ETAPA 2: Somar os leitos ---
# Convertendo para numérico (caso estejam como texto) e tratando valores vazios (NaN) como 0
colunas_leitos = ['QTLEITP1', 'QTLEITP2', 'QTLEITP3']
for col in colunas_leitos:
    df_datasus[col] = pd.to_numeric(df_datasus[col], errors='coerce').fillna(0)

df_datasus['leitos'] = df_datasus['QTLEITP1'] + df_datasus['QTLEITP2'] + df_datasus['QTLEITP3']

# --- ETAPA 3: Unir os arquivos ---
# Criamos um dataframe auxiliar apenas com as colunas necessárias do Datasus para o join
df_leitos_resumo = df_datasus[['CO_CNES', 'leitos']]

# Fazemos o merge (Left Join: mantém tudo das facilidades e adiciona o que encontrar no Datasus)
df_final = pd.merge(df_facilidades, df_leitos_resumo, on='CO_CNES', how='left')

# Preencher com 0 os estabelecimentos que não foram encontrados no arquivo de leitos
df_final['leitos'] = df_final['leitos'].fillna(0).astype(int)

# --- ETAPA 4: Salvar o arquivo ---
df_final.to_csv('Hospitais_gerais_SP_CNES.csv', index=False, encoding='utf-8-sig')

print("Processo concluído! Arquivo de facilidades com leitos gerado.")
```
